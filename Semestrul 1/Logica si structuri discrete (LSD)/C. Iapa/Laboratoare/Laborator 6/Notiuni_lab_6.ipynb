```python
"""
Se considera o lista de liste de siruri de caractere. Sa se scrie in Python o functie cu semnatura
transformare_lista(lista)
care va returna o lista de siruri de caractere ce contin concatenate primele caractere ale fiecarui sir de caractere din fiecare sublista.

Exemplu:
transformare_lista([["Test", "LSD", "Octombrie"], ["AC"], ["Sir", "de", "caractere"]])
=> ["TLO", "A", "Sdc"]

Nota: In cadrul rezolvarii folositi DOAR RECURSIVITATE SI/SAU FUNCTIILE MAP/FILTER/REDUCE
"""
```

```python
from functools import reduce

def reducere(lista):
  return reduce(lambda acc, curr: acc + curr[0],lista , "")

def transformare_lista(listaListe):
  return list(map(reducere, listaListe))

transformare_lista([["Test", "LSD", "Octombrie"], ["AC"], ["Sir", "de", "caractere"]])
```

```python
"""
Se considera o lista de liste de numere intregi. Sa se scrie in Python o functie cu semnatura
transformare_lista(lista)
care va returna o lista de cifre ce contine toate cifrele din cadrul numerelor din subliste, fara insa a avea duplicate.

Exemplu:
transformare_lista([[12, 341, 7], [18], [19, 25, 802, 123]])
=> [1, 2, 3, 4, 7, 8, 9]

Nota: In cadrul rezolvarii folositi DOAR RECURSIVITATE SI/SAU FUNCTIILE MAP/FILTER/REDUCE
"""
```

```python
from functools import reduce

def prelucrare_cifre(lista, numar):
  if numar == 0:
    return lista
  if numar%10 not in lista:
    lista = lista + [numar%10]
  return prelucrare_cifre(lista, numar//10)

def parcurgere_subliste(lista, sublista):
  lista = reduce(lambda acc, el: prelucrare_cifre(acc, el), sublista, lista) # reduce(lambda acc, el: prelucrare_cifre(acc, el), sublista, lista) -> lista
  return lista

def transformare_lista(lista_de_liste):
  return reduce(lambda acc, el: parcurgere_subliste(acc, el), lista_de_liste, [])


transformare_lista([[12, 341, 7], [18], [19, 25, 802, 123]])
```

```python
#Exercitiu 1: Functie care primeste ca parametru un numar natural si returneaza multimea divizorilor. Ex. pentru n = 24 se va returna {2, 3, 4, 6, 8, 12}
def multimea_divizorilor(nr, i=2, multime=None):
  if multime is None:
    multime = set()
  if nr%i == 0:
    multime.add(i)
  elif i > nr/2:
    return multime
  return multimea_divizorilor(nr, i+1, multime)

print(multimea_divizorilor(24))
```

```python
import math

def multimea_divizorilor(nr, i=2, multime=None):
  if multime is None:
    multime = set()
  if nr%i == 0:
    multime.add(i)
    multime.add(nr//i) # adaugam si celalalt divizor
  elif i > math.sqrt(nr):
    return multime
  return multimea_divizorilor(nr, i+1, multime)

print(multimea_divizorilor(24))

from functools import reduce
#cu reduce
def multimea_divizorilor2(nr):
  return reduce(lambda acc, el: acc | {el, nr//el} if nr%el == 0 else acc, range(2, math.ceil(math.sqrt(nr))), {2, nr//2})

print(multimea_divizorilor2(24))
```

```python
"""
Exercitiu 3: Scrieți o funcție care ia o listă de perechi (de tip precizat) și returnează mulțimea elementelor de pe prima poziție din fiecare pereche
(variante: a doua poziție; ambele poziții, dacă sunt de același tip).
Input: [(1,2), (3,4)]; Output: {1,3}
"""
from functools import reduce

def functie(lista_pereche):
  return reduce(lambda acc, el:  acc | {el[0], el[1]}, lista_pereche, set())

print(functie([(1,2), (3,4), (8, 1)]))
```

```python
"""
Implementați funcția standard partition care ia ca parametri o funcție booleană f și o mulțime s și returnează o pereche de mulțimi,
cu elementele din s care satisfac, respectiv nu satisfac funcția f.

Input: lambda x: x % 2 == 0, {1, 2, 3, 4}; Output: ({2, 4}, {1, 3})
"""
from functools import reduce

def partition(multime, conditie):
  rezultat = (set(), set())     # (multimea elementelor care satisfac conditia, multimea elementelor care nu satisfac conditia)
  return reduce(lambda acc, curr: (acc[0] | {curr}, acc[1]) if conditie(curr) else (acc[0], acc[1] | {curr}), multime , rezultat)  

print(partition({1, 2, 3, 4},lambda x: x % 2 == 0 ))
```

Input: lambda x: x % 2 == 0, {1, 2, 3, 4}; Output: {2, 4}

```python
def my_filter(f, multime):
    return reduce(lambda acc, el: acc | {el} if f(el) else acc, multime, set())

print(my_filter(lambda x: x % 2 == 0, {1, 2, 3, 4}))

def filtrare(functie, multime):
    return {x for x in multime if functie(x)}

print(filtrare(lambda x: x % 2 == 0, {1, 2, 3, 4}))
```

4. Implementați funcția standard partition care ia ca parametri o funcție booleană f și o mulțime s și returnează o pereche de mulțimi, cu elementele din s care satisfac, respectiv nu satisfac funcția f.

Input: lambda x: x % 2 == 0, {1, 2, 3, 4}; Output: ({2, 4}, {1, 3})

```python
from functools import reduce

def partition(conditie, multime):
    rezultat = (set(), set())
    return reduce(lambda acc, el: (acc[0] | {el}, acc[1]) if conditie(el) else (acc[0], acc[1] | {el}), multime, rezultat)

print(partition(lambda x: x % 2 == 0, {1, 2, 3, 4}))

def partition2(f,multime):
    multime1 = set()
    multime0 = set()

    for element in multime:
        if f(element):
            multime1.add(element)
        else:
            multime0.add(element)
    return (multime1, multime0)

def partition_lambda(f,multime):
    multime1 = set(filter(lambda x: f(x), multime))
    multime0 = set(filter(lambda x: not f(x), multime))
    return (multime1, multime0)

print(partition2(lambda x: x % 2 == 0, {1, 2, 3, 4}))
```

5. Scrieți o funcție care ia o lista de mulțimi (de exemplu, de șiruri), și returnează reuniunea (variantă: intersectia) mulțimilor.


Input: [{1, 2, 3}, {1, 2, 3, 4}, {3, 5, 6, 7}]; Output: reuniune: {1, 2, 3, 4, 5, 6, 7}; intersectie: {3}

```python
def reuniune_lista(multime):
    return reduce(lambda acc, el: acc | el, multime, set())

print(reuniune_lista([{1, 2, 3}, {1, 2, 3, 4}, {3, 5, 6, 7}]))

def unire_lista(lista_multimi):
    multime = set()
    for element in lista_multimi:
        multime = multime | element
    return multime

print(unire_lista([{1, 2, 3}, {1, 2, 3, 4}, {3, 5, 6, 7}]))

def intersectie_lista(multime):
    return reduce(lambda acc, el: acc & el, multime)

print(intersectie_lista([{1, 2, 3}, {1, 2, 3, 4}, {3, 5, 6, 7}]))
```

7. Scrieți o funcție care ia o mulțime de numere și returnează o listă de perechi (număr, număr de apariții).'

```python
def count_numbers(numbers):
    counts = {}
    for num in numbers:
        if num in counts:
            counts[num] += 1
        else:
            counts[num] = 1
    return list(counts.items())

numbers = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4]
print(count_numbers(numbers))  # Output: [(1, 1), (2, 2), (3, 3), (4, 4)]

from functools import reduce

def count_numbers(numbers):
    def reducer(counts, num):
        if num in counts:
            counts[num] += 1
        else:
            counts[num] = 1
        return counts

    counts = reduce(reducer, numbers, {})
    return list(counts.items())  

numbers = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4]
print(count_numbers(numbers))  # Output: [(1, 1), (2, 2), (3, 3), (4, 4)]

#.count() - numara de cate ori apare un element intr-o lista
print([1, 2, 2, 3, 3, 3, 4, 4, 4, 4].count(4))

```
